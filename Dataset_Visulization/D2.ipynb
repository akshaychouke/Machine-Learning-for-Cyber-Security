```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("../Dataset2.csv")
```

```python
# Box Plot for all features
plt.figure(figsize=(20, 7))
data.boxplot(grid=False)
plt.title("Box Plot for Dataset Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histograms for all features
plt.figure(figsize=(16, 10))
data.hist(bins=20, figsize=(16, 10))
plt.suptitle("Histograms for Dataset Features", y=1.02)
plt.show()
```

```python
# Scatter Matrix for a subset of features (adjust as needed)
from pandas.plotting import scatter_matrix
scatter_matrix(data[['url_length', 'valid_url', 'at_symbol','sensitive_words_count', 'isHttps', 'nb_dots','nb_hyphens', 'nb_and', 'nb_or','nb_www', 'nb_com', 'nb_underscore','target']], alpha=0.5, figsize=(10, 10), diagonal='hist')
plt.suptitle("Scatter Matrix for Selected Features", y=0.92)
plt.show()
```
