```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("../Dataset2.csv")
```

```python
# Box Plot for all features
plt.figure(figsize=(40, 30))
data.boxplot(grid=False)
plt.title("Box Plot for Dataset Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histograms for all features
plt.figure(figsize=(40, 30))
data.hist(bins=20, figsize=(40, 30))
plt.suptitle("Histograms for Dataset Features", y=1.02)
plt.show()
```

```python
# Scatter Matrix for a subset of features (adjust as needed)
from pandas.plotting import scatter_matrix

scatter_matrix(
    data[
        [
            "google_index",
            "page_rank",
            "nb_hyperlinks",
            "web_traffic",
            "nb_www",
            "domain_age",
            "phish_hints",
            "ratio_extHyperlinks",
            "safe_anchor",
            "ratio_intHyperlinks",
            "longest_word_path"
        ]
    ],
    alpha=0.5,
    figsize=(40, 30),
    diagonal="hist",
)
plt.suptitle("Scatter Matrix for Selected Features", y=0.92)
plt.show()
```
