```python
import pandas as pd
import matplotlib.pyplot as plt

# Read your dataset from a CSV file using Pandas
# Replace 'your_dataset.csv' with your actual dataset file
data = pd.read_csv('../Dataset1.csv')

data = data.drop('url_length', axis=1)

# Split the columns into X and Y
num_columns = len(data.columns)
num_columns_half = num_columns // 2

x = data.iloc[:, :num_columns_half]  # The first half of columns
y = data.iloc[:, num_columns_half:]  # The second half of columns

# Create a scatter plot
plt.figure(figsize=(10, 6))
for column in x.columns:
    plt.scatter(x[column], y[column], s=5, alpha=0.5, label=column)

plt.xlabel('X-Axis Label')
plt.ylabel('Y-Axis Label')
plt.title('Scatter Plot of Large Dataset')
plt.legend()

# Show the plot
plt.show()
```
