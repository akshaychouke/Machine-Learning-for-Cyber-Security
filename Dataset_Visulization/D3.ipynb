```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset
data = pd.read_csv("../Dataset3.csv")
```

```python
# Box Plot for all features
plt.figure(figsize=(40, 30))
data.boxplot(grid=False)
plt.title("Box Plot for Dataset Features")
plt.xticks(rotation=45)
plt.show()
```

```python
# Histograms for all features
plt.figure(figsize=(40, 30))
data.hist(bins=20, figsize=(40, 30))
plt.suptitle("Histograms for Dataset Features", y=1.02)
plt.show()
```

```python
# Scatter Matrix for a subset of features
from pandas.plotting import scatter_matrix
scatter_matrix(data[['qty_comma_file', 'time_domain_activation', 'directory_length', 'asn_ip', 'time_response', 'ttl_hostname', 'qty_dot_domain']], alpha=0.5, figsize=(40, 30), diagonal='hist')
plt.suptitle("Scatter Matrix for Selected Features", y=0.92)
plt.show()
```
